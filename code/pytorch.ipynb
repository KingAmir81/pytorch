## Gradient Descent

### 1-D GD

```python
i = 0
error = 5e10000
xa = 20
alpha = 0.05
while abs(error) > 0.01:
  xa = xa - alpha*dfx.subs(x,xa)
  error = dfx.subs(x,xa)
  print(error)
print("-"*20)
print(fx.subs(x,xa))
print(xa)
```

### 2-D GD

```python
y = sym.symbols('y')
z = sym.symbols('z')
fzy = (y)*z**2 - z*y**2 + 2*y*z
dzfzy = sym.diff(fzy,z)
dyfzy = sym.diff(fzy,y)
dzyfzy = sym.diff(fzy,z,y)
```

```python
dzyfzy
```

```python
fzy.evalf(subs={y: 20,z : 20})
```

```python
v = sym.lambdify((y,z),fzy,'numpy')
dvz = sym.lambdify((y,z),dzfzy,'numpy')
dvy = sym.lambdify((y,z),dyfzy,'numpy')
dvzy = sym.lambdify((y,z),dzyfzy,'numpy')
dvy(1,1)
```

```python
dyfzy
```

```python
i = 0
error = 5e10000
ya, za = [1,1]
alpha = 0.05
while abs(error) > .1 :
  dy = dvy(ya,za)
  dz = dvz(ya,za)
  ya = ya - alpha*dy
  za = za - alpha*dz
  error = dvzy(ya,za)
  print(error)
print("-"*20)
print(v(ya,za))
print(ya,za)
```

### ANN Regression
-----

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as display
```

```python
N = 40
m = -2
x = torch.randn(size=(N,1))
y = m*x + torch.randn(size=(N,1))/3
plt.plot(x,y,'ob')
plt.legend(['Real data'])
plt.title('regression')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Model

```python
# build asequential model

ann_reg = nn.Sequential(
    nn.Linear(1,1), # input layer
    nn.ReLU(),      # activation function
    nn.Linear(1,1)  # output layer
)
ann_reg
```

### Meta-parameters

```python
# learning rate
LR = .05
# loss function
lossfun = nn.MSELoss()
# optimizer
optimizer = torch.optim.SGD(ann_reg.parameters(),lr=LR)
```

### training

```python
numepochs = 500
losses_logs = torch.zeros(numepochs)

for ei in range(numepochs):
  yh = ann_reg(x)

  loss = lossfun(yh,y)
  losses_logs[ei] = loss

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```

```python
plt.plot(losses_logs.detach(),'o')
plt.show()
```

### test data
-----

```python
prediction = ann_reg(x)

plt.plot(x,y,'ob')
plt.plot(x,prediction.detach(),'sg')
plt.show()
```

```python
for param in ann_reg.parameters():
  print(param.detach())
```

### Code challenge
-----

```python
def Ann_train(x):
  ann_reg = nn.Sequential(
    nn.Linear(1,1), # input layer
    nn.ReLU(),      # activation function
    nn.Linear(1,1)  # output layer
)
  # learning rate
  LR = .05
  # loss function
  lossfun = nn.MSELoss()
  # optimizer
  optimizer = torch.optim.SGD(ann_reg.parameters(),lr=LR)
  numepochs = 500
  losses_logs = torch.zeros(numepochs)

  for ei in range(numepochs):
    yh = ann_reg(x)

    loss = lossfun(yh,y)
    losses_logs[ei] = loss

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    prediction = ann_reg(x)
  return [losses_logs , prediction]
```

```python
N = 30
l = np.linspace(-2,2,21)

for m in l:
  fig , ax = plt.subplots(1,2,figsize=(15,5))
  x = torch.randn(size=(N,1))
  y = m*x + torch.randn(size=(N,1))

  o  = Ann_train(x)


  ax[0].plot(x,y,'s')
  ax[0].plot(x,o[1].detach(),'og')
  ax[1].plot(o[0].detach(),'og')
  ax[0].set_title(f'm={m} data and regression')
  ax[1].set_title(f'm={m} losses')
  plt.show()
  print("")
  print("")
```
